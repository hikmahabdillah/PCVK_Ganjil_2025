```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://i.pinimg.com/736x/19/de/0d/19de0d946bede94777b16c1dada44124.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                   # read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB
    final_frame = cv.hconcat((image, image_2))       # concatenate image
    cv2_imshow(final_frame)                  # show image
    print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]

print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal ditengah image
for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

[255,255,255] adalah representasi warna putih dalam format RGB.

R = 255 → merah maksimum

G = 255 → hijau maksimum

B = 255 → biru maksimum

5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!

jawab:
- Pixel adalah unit terkecil dari gambar digital (satu titik warna).
- Resolusi adalah jumlah pixel dalam gambar (misalnya 1920x1080 = 2.073.600 pixel).

Resolusi tinggi → lebih banyak pixel → detail gambar lebih tajam, ukuran file lebih besar.
Resolusi rendah → lebih sedikit pixel → gambar tampak buram atau pecah saat diperbesar, ukuran file lebih kecil.

```python
# Ubah format warna gambar dari BGR (default OpenCV) ke RGB
# Disimpan ke image_2 agar gambar asli dalam format RGB tetap ada
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ubah lagi ke RGB untuk salinan yang akan dimodifikasi (image_3)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ambil ukuran gambar: tinggi (jumlah baris piksel) dan lebar (jumlah kolom piksel)
tinggi, lebar = image_3.shape[:2]

# Membuat garis diagonal dari kiri atas → kanan bawah
for i in range(min(tinggi, lebar)):  # Loop sebanyak sisi terkecil (supaya tidak keluar batas)
    image_3[i, i] = [255, 255, 255]  # Ubah piksel diagonal ke warna putih (RGB)

# Membuat garis diagonal dari kiri bawah → kanan atas
for i in range(min(tinggi, lebar)):
    # Akses piksel diagonal dari bawah kiri (y = tinggi-1) naik ke atas
    image_3[tinggi - 1 - i, i] = [255, 255, 255]  # Ubah piksel jadi putih

final_frame_2 = cv.hconcat((image_2, image_3))

# Menampilkan hasil gabungan gambar di layar (cocok untuk Google Colab/Jupyter)
cv2_imshow(final_frame_2)
```

```python
# Ubah format warna gambar dari BGR (default OpenCV) ke RGB
# Disimpan ke image_2 sebagai gambar asli
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ubah lagi ke RGB untuk salinan yang akan dimodifikasi (image_3)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ambil ukuran gambar: tinggi (jumlah baris piksel) dan lebar (jumlah kolom piksel)
tinggi, lebar = image_3.shape[:2]

# Tentukan panjang garis (jumlah piksel) dan posisi awal pada sumbu X
panjang = 100   # garis sepanjang 100 piksel
geser_x = 75    # garis dimulai dari posisi x = 75

# Membuat garis horizontal di tengah gambar
for y in range(panjang):
    # Titik awal: (tinggi/2, geser_x), lalu diperpanjang ke kanan sepanjang 'panjang'
    image_3[int(tinggi/2), geser_x + y] = [255, 255, 255]  # putih

final_frame = cv.hconcat((image_2, image_3))

# Menampilkan hasil gabungan gambar di layar (cocok untuk Google Colab/Jupyter)
cv2_imshow(final_frame)
```

```python
# Ubah format warna gambar dari BGR (default OpenCV) ke RGB
# Disimpan ke image_2 sebagai gambar asli
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ubah lagi ke RGB untuk salinan yang akan dimodifikasi (image_3)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ambil ukuran gambar: tinggi (jumlah baris piksel) dan lebar (jumlah kolom piksel)
tinggi, lebar = image_3.shape[:2]

# Tentukan ukuran kotak
kotak_lebar = 100   # lebar kotak dalam piksel
kotak_tinggi = 100  # tinggi kotak dalam piksel

# Posisi geser untuk kotak
geser_x = 75           # posisi horizontal (jarak dari sisi kiri gambar)
geser_y = tinggi // 2  # posisi vertikal (tengah gambar)

# Hitung koordinat kotak
start_point = (geser_x, geser_y - kotak_tinggi // 2)         # titik kiri atas kotak
end_point   = (geser_x + kotak_lebar, geser_y + kotak_tinggi // 2)  # titik kanan bawah kotak

# Gambar kotak putih di atas image_3
# cv.rectangle(image, start, end, color, thickness)
# - start: koordinat kiri atas
# - end: koordinat kanan bawah
# - color: (255,255,255) → putih
# - thickness: -1 artinya kotak diisi penuh (filled rectangle)
cv.rectangle(image_3, start_point, end_point, (255, 255, 255), -1)

# Menggabungkan dua gambar secara horizontal:
# - image_2 (gambar asli RGB) di kiri
# - image_3 (gambar dengan kotak putih) di kanan
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan hasil gabungan gambar di layar (cocok untuk Google Colab/Jupyter)
cv2_imshow(final_frame)
```
